### Range Function

- range( start,end,step)
- it includes all intgers from `start` up to `end` with an increment of `step` ( including no start but not including end)
- if `step` is not provided then it is 1 by defult

```python
range(5,11)
```

- list command can be used to create list of no generated by range

```python
list(range(5,11))
```

```python
sum = 0
for i in range(5,11):
    sum += i    
sum    
```

```python
list(range(2,19,2))
```

```python
list(range(20,1,-1))
```

```python
for i in range(1,10):
    print(i)
```

```python
import math
for i in range(1,10):
    print(i, math.factorial(i))
```

### Scientific No approximation
- for too large float no and too small float no scientific no system is used to approximate the value
- it will help to pravent memory overfloe


```python
1.156e+5
```

```python
import math
for i in range(1,40,3):
    print(i, float(math.factorial(i)))
```
